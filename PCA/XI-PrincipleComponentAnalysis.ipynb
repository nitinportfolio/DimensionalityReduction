PCA - Bring high dimensional data into lower dimensional data keeping the escense of data intact.
eg. Moves are shoot in 3 dimensions using camera but we watch them in 2 dimensions without loosing any required information.
Benefits - Faster execution as number of features are reduced & easy visualization


```python
import numpy as np
import pandas as pd

np.random.seed(101)

my_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])

class1_sample = np.random.multivariate_normal(my_vec1, cov_mat1, 20)
```

Using algebra for PCA we will bring 3d into 2d with best 2 dimensions.

```python
# Step 1 - Apply standard scalling - Mean centric
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Mean centric
df.iloc[:,0:3] = scaler.fit_transform(df.iloc[:,0:3])
```

```python
df
```

```python
# Step 2 - Find the covariance Matrix
```

```python
covariance_matrix = np.cov([df.iloc[:,0], df.iloc[:,1], df.iloc[:,2]])
print("Covariance Matrix :\n", covariance_matrix)
```

```python
# In above fig. diagonals are variances of three vectors & non diagoanls are covariances
```

```python
# Step 3 finding Eign values & vectors
# using linear algebra eigen np.linalg.eig
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python

%matplotlib inline

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
from matplotlib.patches import FancyArrowPatch


class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        super().__init__((0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def do_3d_projection(self, renderer=None):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, self.axes.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))

        return np.min(zs)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df['Feature1'], df['Feature2'], df['Feature3'], 'o', markersize=8, color='blue', alpha=0.2)
ax.plot([df['Feature1'].mean()], [df['Feature2'].mean()], [df['Feature3'].mean()], 'o', markersize=10, color='red', alpha=0.5)
for v in eigen_vectors.T:
    a = Arrow3D([df['Feature1'].mean(), v[0]], [df['Feature2'].mean(), v[1]], [df['Feature3'].mean(), v[2]], mutation_scale=20, lw=3, arrowstyle="-|>", color="r")
    ax.add_artist(a)
ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_zlabel('z_values')

plt.title('Eigenvectors')

plt.show()
     
```

```python
# Fetchi ng top two eigen vectors
pc = eigen_vectors[0:2]
pc
```

```python

transformed_df = np.dot(df.iloc[:,0:3],pc.T)
# 40,3 - 3,2
new_df = pd.DataFrame(transformed_df,columns=['PC1','PC2'])
new_df['target'] = df['target'].values
new_df.head()
```

```python

new_df['target'] = new_df['target'].astype('str')
fig = px.scatter(x=new_df['PC1'],
                 y=new_df['PC2'],
                 color=new_df['target'],
                 color_discrete_sequence=px.colors.qualitative.G10
                )

fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
     
```
